```python
plt.figure(figsize=(6,6))
sns.histplot(dataset['SalePrice'],kde=True, bins=30)
plt.title('Distribution of Sale Prices')
plt.xlabel('Sale price')
plt.ylabel('Number of Houses')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(dataset.corr(numeric_only=True),cmap='Paired',fmt='.2f',linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

```python
plt.scatter (x='SalePrice', y='GrLivArea', data=dataset)
plt.title('Relationship between Ground Living Area (GeLivArea) and Sale Price')
plt.xlabel('Sale Price')
plt.ylabel('Ground Living Area')
plt.show()
```

```python
dataset['AgeOfHouse'] = dataset['YrSold'] - dataset['YearBuilt']
```

```python
dataset.head(5)
```

```python
plt.scatter(x='SalePrice', y='AgeOfHouse', data = dataset)
plt.title("Relationship between SalaPrice and AgeOfHouse")
plt.xlabel('Sale Price')
plt.ylabel('Age of House')
plt.show()
```

```python
dataset['TotalUsableArea'] = dataset['GrLivArea'] + dataset['GarageArea']
dataset['PricePerTotalSqft'] = dataset['SalePrice']/dataset['TotalUsableArea']
```

```python
dataset.sample(5)
```

```python
corr = dataset.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=corr.head(10), y=corr.head(10).index,palette='viridis')
plt.title('Top Features Correlated with SalePrice')
plt.xlabel('Correlation Coefficint')
plt.ylabel('Total Useble Area')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(dataset['OverallQual'],dataset['TotalUsableArea'], c=dataset['SalePrice'],cmap='viridis')
plt.colorbar(label='Sale Price')
plt.xlabel('Overall Quality')
plt.ylabel('Total Usable Area')
plt.title('Overall Quality vs Total Usable Area (Coloured by Sale Price)')
plt.show()
```

```python
sns.boxplot(x=dataset['PricePerTotalSqft'])
plt.title('Boxplot of Price Per Total Squre Foot')
plt.xlabel('Price Per Sqft')
plt.show()
```
